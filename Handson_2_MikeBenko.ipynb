## 3.1 Letter Grade Converter
Create a program that converts number grades to letter grades.

### Console:
```powershell
Letter Grade Converter

Enter numerical grade: 90 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 88 
Letter grade: A

Continue? (y/n): y

Enter numerical grade: 80 
Letter grade: B

Continue? (y/n): y

Enter numerical grade: 67 
Letter grade: C

Continue? (y/n): y

Enter numerical grade: 59 
Letter grade: F

Continue? (y/n): n

Bye!
```

### Specifications:
- Assume that the user will enter valid integers for the grades.
- The program should continue only if the user enters “y” or “Y” to continue.
- The grading criteria is as follows:

| Letter |	Range |
|---|-------|
| A |	88-100 |
| B |	80-87 |
| C |	67-79 |
| D |	60-66 |
| F |	<60 |

````python
i = 'y'

while i.lower() == 'y':
    
    grade = int(input("Please enter numerical grade: "))
    if grade > 100:
        print("Please enter a valid numerical grade.")
    elif grade >= 88:
        print("You earned a letter grade of: A")
    elif grade >= 80:
        print("You earned a letter grade of: B")
    elif grade >= 67:
        print("You earned a letter grade of: C")
    elif grade >= 60:
        print("You earned a letter grade of: D")
    elif grade >= 0:
        print("You earned a letter grade of: F")
    else:
        print("Please enter a valid numerical grade.")

    i = input("Continue? (y/n): ")
````

## 3.2 - Tip Calculator
Create a program that calculates three options for an appropriate tip to leave after a meal at a restaurant.

### Console
```powershell
Tip Calculator

Cost of meal: 52.31 

15%
Tip amount:	7.85
Total amount: 60.16

20%
Tip amount:	10.46
Total amount: 62.77

25%
Tip amount:	13.08
Total amount: 65.39
```

### Specifications:
- The program should calculate and display the cost of tipping at 15%, 20%, or 25%.
- Assume the user will enter valid data.
- The program should round results to a maximum of two decimal places.



````python
i = 'y'
# initial while loop to utilize continue condition
while i.lower() == 'y':
    meal_cost = float(input("Enter the total cost of your meal: (X.XX)"))
    calc_tip = .15
    print("\nTip Calculator")
    print(f"\nCost of Meal: ${meal_cost:.2f}")
    print()

    # while loop to interate through each tip % level
    while calc_tip < .26:
        print(str(calc_tip * 100) + "%")
        tip_out = meal_cost * calc_tip          # calc tip total
        total = meal_cost + tip_out             # add tip total & meal cost to get total
        print(f"Tip amount:   {tip_out:.2f}")   # formatted outputs using variable call
        print(f"Total amount: {total:.2f}")
        calc_tip = 0.05 + calc_tip
        # simple double space to seperate the tip calculations
        print("\n")
    # finally ask user for cont condition    
    i = input("Continue? (y/n): ")
````

## 3.3 - Change Calculator
Create a program that calculates the coins needed to make change for the specified number of cents.

### Console
```powershell
Change Calculator

Enter number of cents (0-99): 99

Quarters: 3
Dimes:	2
Nickels: 0
Pennies: 4

Continue? (y/n): y

Enter number of cents (0-99): 55

Quarters: 2
Dimes:	0
Nickels: 1
Pennies: 0

Continue? (y/n): n 

Bye!
```

### Specifications
- The program should display the minimum number of quarters, dimes, nickels, and pennies that one needs to make up the specified number of cents.
- Assume that the user will enter a valid integer for the number of cents.
- The program should continue only if the user enters “y” or “Y” to continue.


````python
i = 'y'


while i.lower() == 'y':
    quarters = 0
    dimes = 0
    nickels = 0
    pennies = 0
    print("Change Calculator")
    cents = int(input("\nEnter number of cents (0-99): "))
    while cents >= 0:
        if cents >= 25:
            quarters+=1
            cents = cents - 25
            continue
        elif cents >= 10:
            dimes+=1
            cents = cents - 10
            continue
        elif cents >= 5:
            nickels+=1
            cents = cents - 5
            continue
        elif cents >= 1:
            pennies+=1
            cents = cents - 1
            continue
        else:
            break
    print(f"Quarters: {quarters}")
    print(f"Dimes:    {dimes}")
    print(f"Nickels:  {nickels}")
    print(f"Pennies:  {pennies}")
    print()
    i = input("Continue? (y/n): ")
if i.lower() == 'n':
    print("Bye!")

````

## 3.4 - Table of Powers
Create a program that displays a table of squares and cubes for the specified range of numbers.

### Console
```powershell
Table of Powers

Start number: 90
Stop number: 100

Number	Squared	Cubed
======	=======	=====
90      8100	729000
91      8281	753571
92      8464	778688
93      8649	804357
94      8836	830584
95      9025	857375
96      9216	884736
97      9409	912673
98      9604	941192
99      9801	970299
100     10000	1000000
```

### Specifications
- The formulas for calculating squares and cubes are:

    `square = x ** 2`
    
    `cube = x ** 3`

- Use tabs to align the columns.
- Assume that the user will enter valid integers.
- Make sure the user enters a start integer that’s less than the stop integer. If the user enters a start integer that’s greater than the stop integer, display an error message and give the user a chance to enter the integers again.


````python
print("Table of Powers")
start = int(input("Start Number: "))
stop = int(input("Stop Number: "))
if start > stop:
    start, stop = stop, start

print("Number\tSquared\tCubed")
print("======\t======\t=====")
for x in range (start, stop + 1):
    square = x ** 2
    cube = x ** 3
    print(f"{x}\t{square}\t{cube}")
    
````
